```python
import scqubits as scq
import qutip as qt
import numpy as np
```

```python
m = 60 # number of transmon levels
#Ej = 2.93556 #/2/pi
Ej = 4.0
#Ec = 0.73032 #/2/pi
Ec = 1.0 #/2/pi
#El = 0.55097 #/2/pi
El = 0.75 #/2/pi
phi_ext = 0.5 #*2*pi
flx = scq.Fluxonium(Ej,Ec,El,phi_ext,cutoff = m)
```

```python
x = Ej/Ec
print(x)
x = El/Ec
print(x)
```

```python
evals, evecs = flx.eigensys(evals_count=200)
```

```python
evals, evecs = flx.eigensys(evals_count=60)
```

```python
evals.shape
# (20,)

len(evecs)
# 20
```

```python
N = flx.n_operator()
```

```python
V = np.column_stack(
    [state.flatten() for state in evecs]
)
```

```python
N_eig = V.conj().T @ N @ V
```

```python
np.abs(N_eig[:20,:5])
```

```python
np.abs(N_eig[:20,:20])
```

```python
N = len(evals)
n_minus= np.zeros((N,N), dtype = complex)
for i in range(N):
    for j in range(i+1 , N):
        n_minus[i,j] = N_eig[i,j]

n_plus = n_minus.conj().T
#print(np.abs(n_plus[:20, :20]))
print(np.abs(n_minus[:20, :20]))
np.allclose(N_eig, n_plus + n_minus)
```
